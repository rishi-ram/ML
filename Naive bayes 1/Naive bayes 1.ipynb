```python
df=pd.read_csv('titanic.csv')
target=df.Survived
```

```python
target
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies
```

```python
merge_input=pd.concat([inputs,dummies],axis='columns')
merge_input
```

```python
merge_input.drop('Sex',axis='columns',inplace=True)
merge_input
```

```python
inputs.columns[inputs.isna().any()]
```

```python
merge_input.Age=df.Age
merge_input
```

```python
merge_input.Age=merge_input.Age.fillna(df.Age.mean())
merge_input
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(merge_input,target,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
